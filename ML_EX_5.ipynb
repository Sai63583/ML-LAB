```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'x':[2,2,5,4,8,6,5,5,6,4],'y':[4,6,6,7,3,6,2,7,3,4]})
```

```python
from sklearn.cluster import KMeans
```

```python
x=df
x.head()
```

```python
km=KMeans(n_clusters=3)
km.fit(x)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
df['k mean paradicted label']=km.labels_
labels_=km.labels_
df.head()
```

```python
plt.scatter(df['x'],df['y'],50,labels_)
for i in km.cluster_centers_:
  plt.scatter(i[0],i[1],200,marker='*')
```
